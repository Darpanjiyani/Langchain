# **CHAINS in Langchain**

```python
#from langchain_anthropic import ChatAnthropic
from langchain_openai import ChatOpenAI
import os
from load_dotenv import load_dotenv

load_dotenv() #This function will load everything which is saved in our .env file and will make them available
              # in the os.environ dictionary (env variables)

if os.environ.get("OPENAI_API_KEY"):
    print("Bro, API KEY Variable Exists")
else: 
    raise ValueError("OPENAI_API_KEY not found")

from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_openai = ChatOpenAI(model ='gpt-5-mini', temperature=0)
```

# **First Chain**

```python
# TASK1 - Prompt

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])
```

```python
# TASK2 - LLM

llm_openai = ChatOpenAI(model ='gpt-5-mini', temperature=0)
```

```python
# TASK3 - Str Parser

str_parser = StrOutputParser()
```

# **Manual Invocation**

```python
template = prompt_template.invoke({"input": "What is the capital of France?"})

res = llm_openai.invoke(template)

final_result = res.content

final_result
```

# **Chain Invocation**

1st way to run chains using Pipe

```python
chain = prompt_template | llm_openai | str_parser

chain.invoke({"input": "What is the capital of France?"})
```

2nd type to run chains using RunnableSequence

```python
from langchain_core.runnables import RunnableSequence

chain_1 = RunnableSequence(prompt_template, llm_openai, str_parser)
```
