```python
#from langchain_anthropic import ChatAnthropic
from langchain_openai import ChatOpenAI
import os
from load_dotenv import load_dotenv

load_dotenv() #This function will load everything which is saved in our .env file and will make them available
              # in the os.environ dictionary (env variables)

if os.environ.get("OPENAI_API_KEY"):
    print("Bro, API KEY Variable Exists")
else:
    raise ValueError("OPENAI_API_KEY not found")
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_openai import ChatOpenAI

llm_openai = ChatOpenAI(model ='gpt-5-mini', temperature=0) #temperature is basically creativity of the model, if we setup to 0
                                                            #then model will not be creative. Model will simply generate the answer
                                                            #with best possibility. But if we set let's say 0.9, 0.8, model will just play around the output and it can just generate some random output.

my_message = [
    #SystemMessage(content = "You are a gen-z assistant who always answers in a fun way"),
    SystemMessage(content = "You are a strict minded guy and very angry right now"), 
    HumanMessage(content = "Bro, tell me a fun fact")
]

llm_openai.invoke(my_message).content
```

```python
my_new_message = [
    #SystemMessage(content = "You are a gen-z assistant who always answers in a fun way"),
    SystemMessage(content = "You are a data engineer expert"), 
    HumanMessage(content = "Explain ETL vs ELT")
]

llm_openai.invoke(my_new_message).content
```

# PROMPTS

Prompts are messages that are sent to the LLM.

Prompts are more user-friendly than messages.

```python
from langchain_core.prompts import PromptTemplate

user_input = input("Enter a topic for fun fact : ")

dynamic_prompt = PromptTemplate.from_template("Write a fun fact about {topic}")

dynamic_prompt.invoke({"topic" : user_input})
```

```python
from langchain_core.prompts import PromptTemplate

user_input = input("Enter a topic for fun fact : ")

dynamic_prompt = PromptTemplate.from_template("Write a fun fact about {topic}")

ready_prompt = dynamic_prompt.invoke({"topic" : user_input})

llm_openai.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a polite assistant"),
    ("user", "Write a fun fact about {topic}")
])

ready_prompt = prompt_template.invoke({"topic": "AI"})

ready_prompt.messages
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a polite assistant"),
    ("user", "Write a fun fact about {topic}")
])

user_input = input("Enter a topic: ")

ready_prompt = prompt_template.invoke({"topic": user_input})

llm_openai.invoke(ready_prompt.messages).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {tone} assistant"),
    ("user", "Write a fun fact about {topic}")
])

user_input = input("Enter a topic: ")
user_tone = input('Enter a tone : ')

ready_prompt = prompt_template.invoke({"topic": user_input, "tone": user_tone})

llm_openai.invoke(ready_prompt.messages).content
```
